```
# Calculating average price over last N days of the stock price
def avg(n):
  return df['Close Price'].tail(n).mean()
```

```
n=int(input("Enter the No. of days : "))
avg(n)
```

```
def profit_and_loss(n):
    k = len(df)
    curr = df['Close Price'][k-1]
    n_old_day = df['Close Price'].tail(n)
    
    return ((curr - n_old_day)/n_old_day)*100
```

```
sum(profit_and_loss(7))
# calculating for 1 week
```

```
sum(profit_and_loss(14))
# calculating for 2 weeks
```

```
sum(profit_and_loss(30))
# calculating for 1 month
```

```
sum(profit_and_loss(90))
# calculating for 3 months
```

```
sum(profit_and_loss(180))
# calculating for 6 months
```

```
sum(profit_and_loss(365))
# calculating for 1 year
```

```
df.loc[df['Day_Perc_Change'].between(-0.005, -0.01, inclusive = True), 'Trend'] = 'Slight negative'

df.loc[df['Day_Perc_Change'].between(0.005, 0.01, inclusive = True), 'Trend'] = 'Slight positive'
```

```
df.loc[df['Day_Perc_Change'].between(1, 3, inclusive= True), 'Trend'] = 'Positive'

df.loc[df['Day_Perc_Change'].between(-1, -3, inclusive= True), 'Trend'] = 'Negative'
```

```
df.loc[df['Day_Perc_Change'].between(3, 7, inclusive=True), 'Trend'] = 'Among Top Gainers'

df.loc[df['Day_Perc_Change'].between(-7, -3, inclusive=True), 'Trend'] = 'Among Top Losers'
```
